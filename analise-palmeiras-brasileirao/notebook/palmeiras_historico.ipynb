```python
import pandas as pd

df_2025 = pd.read_csv("palmeiras_jogos_2025.csv")
df_2026 = pd.read_csv("palmeiras_jogos_2026.csv")
```

```python
df_historico = pd.concat(
    [df_2025, df_2026],
    ignore_index=True
)
```

```python
df_historico.to_csv(
    "palmeiras_jogos_historico.csv",
    index=False
)
```

```python
df_casa_fora_2025 = pd.read_csv("palmeiras_casa_fora_2025.csv")
df_casa_fora_2026 = pd.read_csv("palmeiras_casa_fora_2026.csv")
```

```python
df_casa_fora_2025["temporada"] = 2025
df_casa_fora_2026["temporada"] = 2026
```

```python
df_casa_fora_historico = pd.concat(
    [df_casa_fora_2025, df_casa_fora_2026],
    ignore_index=True
)
```

```python
df_casa_fora_historico.to_csv(
    "palmeiras_casa_fora_historico.csv",
    index=False
)
```
